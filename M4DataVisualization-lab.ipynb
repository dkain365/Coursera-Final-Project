# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

# Now, plotting the histogram for 'ConvertedComp'
plt.hist(df['ConvertedComp'].dropna(), bins=10, color='blue', edgecolor='black')
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

plt.boxplot(df['Age'].dropna(), patch_artist=True, boxprops=dict(facecolor='blue'))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.xticks([1], ['Age'])
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.3, color='blue')  
plt.title('Scatter Plot of Age vs. Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

df_clean = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

# Normalize 'Age' for bubble size to ensure it's visually proportional and not too large or small
age_normalized = (df_clean['Age'] - df_clean['Age'].min()) / (df_clean['Age'].max() - df_clean['Age'].min())

# Create the bubble plot
plt.scatter(df_clean['WorkWeekHrs'], df_clean['CodeRevHrs'], s=age_normalized * 1000, alpha=0.5, color='blue')
plt.title('Bubble Plot - Work Week Hours vs. Code Review Hours')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY, conn)


medians = df.groupby('Age').median()

medians.plot(kind='bar', stacked=True, color=['skyblue', 'lightgreen'])
plt.title('Median Work Week Hours and Code Review Hours by Age (30-35)')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.legend(['Work Week Hours', 'Code Review Hours'])

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(QUERY, conn)

# Calculate the median 'ConvertedComp' for each age
medians = df.groupby('Age')['ConvertedComp'].median()


medians.plot(kind='line', marker='o', color='green')
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) AS Frequency
FROM master
GROUP BY MainBranch
ORDER BY Frequency DESC
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(10, 8))  # You can adjust these dimensions as needed


plt.barh(df['MainBranch'], df['Frequency'], color='blue')
plt.xlabel('Frequency')
plt.title('Frequency of Respondents by Main Branch')
plt.tight_layout()  
plt.show()
```
